```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Importing the pandas library and using the abbreviation "pd" for it.
# Importing the LabelEncoder class from the sklearn.preprocessing module.

# Reading the CSV file into a DataFrame.
df = pd.read_csv('/content/IRIS[1].csv')

# Creating a new column for numeric representations of the categorical values.
df['Species_Numeric'] = LabelEncoder().fit_transform(df['species'])

# Displaying the updated DataFrame.
df
```

```python
del df['species']

# Removes the 'species' column from the DataFrame.

df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))

# Importing the seaborn library and using the abbreviation "sns" for it.
# Calling plt.figure() function to create a figure for the plot.
# Setting the figure size with the figsize parameter.

sns.heatmap(df.corr(), annot=True, cmap="RdYlGn")

# Calculating the correlation between columns in the DataFrame using df.corr() function.
# Displaying the correlation values on the heatmap with annot=True parameter.
# Setting the colormap to "RdYlGn" with cmap parameter.
```

```python
x = df[['sepal_length', 'petal_length', 'petal_width']]

# Creates an 'x' variable containing the columns 'sepal_length', 'petal_length', and 'petal_width'.

y = df['Species_Numeric']

# Creates a 'y' variable containing the 'Species_Numeric' column.
```

```python
from sklearn.model_selection import train_test_split

# Importing the train_test_split function from sklearn.model_selection.

xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size=0.3, random_state=1)

# Using the train_test_split function to split the data into training and testing sets.
# 'x' represents the DataFrame containing the independent variables.
# 'y' represents the Series containing the dependent variable.
# test_size=0.3 parameter specifies that 30% of the data will be used for the testing set.
# random_state=1 parameter is used to fix the randomness (for reproducibility).
# xTrain, xTest, yTrain, yTest are created as the training and testing sets, respectively.
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Importing the LogisticRegression class from the sklearn.linear_model module.
# Importing the accuracy_score and confusion_matrix functions from the sklearn.metrics module.
# Importing the pandas library and using the abbreviation "pd" for it.
# Importing the seaborn library and using the abbreviation "sns" for it.
# Importing the matplotlib.pyplot library and using the abbreviation "plt" for it.

logR = LogisticRegression()
logR.fit(xTrain, yTrain)

# Creating a 'logR' object of the LogisticRegression class.
# Fitting the model using the fit() function with the training data.

logRpredict = logR.predict(xTest)

# Making predictions on the test data using the predict() function.

logRaccuracy_scoree = accuracy_score(yTest, logRpredict)

# Calculating the accuracy score using the accuracy_score() function.

logRCMscore = confusion_matrix(yTest, logRpredict)

# Calculating the confusion matrix using the confusion_matrix() function.

cm2 = pd.DataFrame(logRCMscore, columns=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'],
                   index=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])

# Converting the confusion matrix to a DataFrame for better visualization.

print("accuracy_score score LogisticRegression = ", end='')
print(logRaccuracy_scoree)

# Printing the accuracy score of the Logistic Regression model.

hm = sns.heatmap(cm2, annot=True, cmap="YlOrRd", fmt='n')

# Creating a heatmap using seaborn to visualize the confusion matrix.
# annot=True parameter shows the matrix values on the heatmap.
# cmap="YlOrRd" parameter sets the colormap for the heatmap.
# fmt='n' parameter shows the matrix values as numbers.

hm.tick_params(labeltop=True, labelbottom=False)
hm.set_xlabel('PREDICTED (LogisticRegression)')
hm.xaxis.set_label_position('top')
hm.set_ylabel('ACTUAL')
# Setting the axis labels and positions.

plt.show()
# Displaying the plot.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import pandas as pd

# Importing the seaborn library and using the abbreviation "sns" for it.
# Importing the matplotlib.pyplot library and using the abbreviation "plt" for it.
# Importing the DecisionTreeClassifier class from the sklearn.tree module.
# Importing the accuracy_score and confusion_matrix functions from the sklearn.metrics module.
# Importing the pandas library and using the abbreviation "pd" for it.

DTC = DecisionTreeClassifier()

# Creating a 'DTC' object of the DecisionTreeClassifier class.

DTC.fit(xTrain, yTrain)
DTCpredict = DTC.predict(xTest)

# Fitting the model using the fit() function with the training data.
# Making predictions on the test data using the predict() function.

DTFaccuracy_scoree = accuracy_score(yTest, DTCpredict)

# Calculating the accuracy score using the accuracy_score() function.

DTFCMscore = confusion_matrix(yTest, DTCpredict)

# Calculating the confusion matrix using the confusion_matrix() function.

cm2 = pd.DataFrame(DTFCMscore, columns=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'],
                   index=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])

# Converting the confusion matrix to a DataFrame for better visualization.

print("accuracy_score score DecisionTreeClassifier = ")
print(DTFaccuracy_scoree)

# Printing the accuracy score of the Decision Tree Classifier model.

hm = sns.heatmap(cm2, annot=True, cmap="YlOrRd", fmt='n')

# Creating a heatmap using seaborn to visualize the confusion matrix.
# annot=True parameter shows the matrix values on the heatmap.
# cmap="YlOrRd" parameter sets the colormap for the heatmap.
# fmt='n' parameter shows the matrix values as numbers.

hm.tick_params(labeltop=True, labelbottom=False)
hm.set_xlabel('PREDICTED (DecisionTreeClassifier)')
hm.xaxis.set_label_position('top')
hm.set_ylabel('ACTUAL')

# Setting the axis labels and positions.

plt.show()

# Displaying the plot.
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Importing the KNeighborsClassifier class from the sklearn.neighbors module.
# Importing the confusion_matrix and accuracy_score functions from the sklearn.metrics module.
# Importing the pandas library and using the abbreviation "pd" for it.
# Importing the seaborn library and using the abbreviation "sns" for it.
# Importing the matplotlib.pyplot library and using the abbreviation "plt" for it.

kNN = KNeighborsClassifier()
kNN.fit(xTrain, yTrain)

# Creating a 'kNN' object of the KNeighborsClassifier class.
# Fitting the model using the fit() function with the training data.

kNN_predict = kNN.predict(xTest)

# Making predictions on the test data using the predict() function.

kNN_accuracy_score = accuracy_score(yTest, kNN_predict)

# Calculating the accuracy score using the accuracy_score() function.

kNN_CM_score = confusion_matrix(yTest, kNN_predict)

# Calculating the confusion matrix using the confusion_matrix() function.

cm2 = pd.DataFrame(kNN_CM_score, columns=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'],
                   index=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])

# Converting the confusion matrix to a DataFrame for better visualization.

print("Accuracy score k-Nearest Neighbors (k-NN): ", end='')
print(kNN_accuracy_score)

# Printing the accuracy score of the k-Nearest Neighbors (k-NN) model.

hm = sns.heatmap(cm2, annot=True, cmap="YlOrRd", fmt='n')

# Creating a heatmap using seaborn to visualize the confusion matrix.
# annot=True parameter shows the matrix values on the heatmap.
# cmap="YlOrRd" parameter sets the colormap for the heatmap.
# fmt='n' parameter shows the matrix values as numbers.

hm.tick_params(labeltop=True, labelbottom=False)
hm.set_xlabel('PREDICTED (KNeighborsClassifier)')
hm.xaxis.set_label_position('top')
hm.set_ylabel('ACTUAL')

# Setting the axis labels and positions.

plt.show()

# Displaying the plot.
```
